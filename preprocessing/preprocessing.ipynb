```python
import os
import stltovoxel

# Configurations
directory_stl_files = 'G://Project/icdas_classification/dataset/ICDAS_FINAL'
directory_npy_files = './npy_files'
directory_sliced = './sliced'

# Create All Paths if they don't exist
if not os.path.exists(directory_stl_files): os.mkdir(directory_stl_files)
if not os.path.exists(directory_npy_files): os.mkdir(directory_npy_files)
if not os.path.exists(directory_sliced): os.mkdir(directory_sliced)

allFiles = os.listdir(directory_stl_files)
for fileIndex, filenameWithExtension in enumerate(allFiles):
    print("Processing: " + filenameWithExtension)
    f = os.path.join(directory_stl_files, filenameWithExtension)
    if os.path.isfile(f) and filenameWithExtension.endswith('.stl'):
        filename = filenameWithExtension.rsplit(".", 1)[0]
        
        # Create Folder if not created
        if not os.path.exists(directory_sliced+'/'+filename): os.mkdir(directory_sliced+'/'+filename)

        # Run STL-TO-VOXEL
        stltovoxel.convert_file(directory_stl_files+'/'+filenameWithExtension,directory_sliced+'/'+filename+'/'+'slice.png',resolution=70)
        # stltovoxel.convert_file(directory_stl_files+'/'+filenameWithExtension,directory_npy_files+'/'+filename+'_'+'slice.npy',resolution=70)
        
```

```python
from PIL import Image
import os

# Define the directory where the PNG images are located
png_dir = './sliced'

# Iterate through each directory
for directory in os.listdir(png_dir):
    # Iterate through each PNG image
    for png in os.listdir(os.path.join(png_dir, directory)):
        # Check if the file is a PNG image
        if png.endswith('.png'):
            # Set the path to the PNG image
            png_path = os.path.join(png_dir, directory, png)
            
            # Open the image using PIL
            img = Image.open(png_path)
            
            # Resize the image to 70x70
            resized_img = img.resize((70, 70))
            
            # Save the resized image
            resized_img.save(png_path)
```

```python
# import the necessary packages
import numpy as np
from PIL import Image
import os
## Define the directory where the PNG images are located
png_dir = './sliced'

## Define the directory where the NPY files will be saved
npy_file = './npy_files'

## Create directory if not created
if not os.path.exists(npy_file): os.mkdir(npy_file)



## get all directories in the directory 
directories = os.listdir(png_dir)

## Loop through each directory
for directory in directories:
    
    ## Create an empty list to hold the image arrays
    image_arrays = []
    
    ## Loop through each PNG image and convert it to a NumPy array
    for png in os.listdir(os.path.join(png_dir, directory)):
        
        ## Set the path to the PNG image
        png_path = os.path.join(png_dir, directory, png)
        
        ## Load the PNG image using PIL
        img = Image.open(png_path)

        ## Convert the image to a NumPy array
        img_array = np.array(img)
        
        ## Change the data type to float32
        img_array = img_array.astype('float32')
        
        ## Print the array
        print(img_array)

        ## Add the NumPy array to the list of image arrays
        image_arrays.append(img_array)
        
        ## Stack the image arrays into a single array
        stacked_array = np.stack(image_arrays, axis=0)
        
        ## Save the stacked array as a NPY file
        save_path = os.path.join(npy_file, directory + '.npy')
        np.save(save_path, stacked_array)
        
```
